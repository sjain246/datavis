```python
#Import Statements
import numpy as np
import pandas as pd
from PIL import Image
```

```python
#Read in the Data
df = pd.read_csv('data.csv')
```

```python
#Ensuring 50 Images Per Class
df['Class'].value_counts()
```

```python
#Example Image with Class and Sentence
loaded_arr = np.load(df.loc[150,'Image Array'])
display(Image.fromarray(loaded_arr))
display('Class: ' + df.loc[150,'Class'])
display('Sentence: ' + df.loc[150,'Sentence'])
```

```python
#Function to display all data for the specified class
def disp_class(cls):
    class_df = df[df['Class'] == cls]
    for i, r in class_df.iterrows():
        loaded_arr = np.load(r['Image Array'])
        display(Image.fromarray(loaded_arr))
        display('Class: ' + r['Class'])
        display('Sentence: ' + r['Sentence'])
```

```python
disp_class('Pizza')
```

```python
disp_class('Airplane')
```

```python
disp_class('Human')
```

```python
disp_class('Elephant')
```
